```python
# Mass produce all the plots, again, but this time the better way
def plot_cartesian_data(pts, cartesian=False):
    positions = []
    common_ax = None
    for point_num, point in enumerate(pts):
        if not cartesian:
            cartesian_data = []
            for trial_num, (dists, angles) in enumerate(point):
                # Convert polar data to Cartesian data
                xs = np.cos(angles) * dists
                ys = np.sin(angles) * dists
                cartesian_data.append([xs, ys])

            # Generate Cartesian scatterplot
            df = pd.DataFrame({
                "x1": cartesian_data[0][0], 
                "y1": cartesian_data[0][1], 
                "x2": cartesian_data[1][0], 
                "y2": cartesian_data[1][1],
            }, )
            positions.append((df, p_to_i[point_num+1]))
        
            # Use same scatterplot for both trial data
            common_ax = df.plot(
                x="x1", 
                y="y1", 
                kind="scatter", 
                title=f"Point {point_num+1}: {repr(p_to_i[point_num+1])}", 
                label="Trial 1", 
                color="Blue"
            )
            df.plot(x="x2", y="y2", kind="scatter", label="Trial 2", color="Red", ax=common_ax)

        else:
            assert type(point) == pd.DataFrame, f"Was expecting Pandas DataFrame, but got {type(point)} instead"
            if common_ax is None:
                common_ax = point.plot(
                    x="x1",
                    y="y1",
                    kind="scatter",
                    title=f"Complete mapping of all accumulated points",
                    label="Trial 1",
                    color="Blue"
                )
            else:
                point.plot(x="x1", y="y1", kind="scatter", label="Trial 1", color="Blue", ax=common_ax)
            point.plot(x="x2", y="y2", kind="scatter", label="Trial 2", color="Red", ax=common_ax)
            
            legend = common_ax.get_legend()
            if legend is not None:
                legend.remove()
    plt.show()

    return positions

positions = plot_cartesian_data(points, cartesian=False)
```

```python
# Process the Cartesian position data to an absolute origin coordinate frame
new_points = []
for pos, (x_orig, y_orig) in positions:
    # Conversion factor from ft -> mm
    x_orig_mm = x_orig * 304.8
    y_orig_mm = y_orig * 304.8
    
    # Add the appropriate mm offsets
    pos["x1"] += x_orig_mm
    pos["x2"] += x_orig_mm
    pos["y1"] += y_orig_mm
    pos["y2"] += y_orig_mm
    
    # Additional filtering - experimental
    x1_bounds = (-2000 < pos["x1"]) & (pos["x1"] < 2500)
    x2_bounds = (-2000 < pos["x2"]) & (pos["x2"] < 2500)
    y1_bounds = (-2000 < pos["y1"]) & (pos["y1"] < 2000)
    y2_bounds = (-2000 < pos["y2"]) & (pos["y2"] < 2000)
    # print(pos[x_bounds][y_bounds])
    
    # Add the new points data to a new array (for plotting later)
    new_points.append(pos[x1_bounds][x2_bounds][y1_bounds][y2_bounds])
    # new_points.append(pos)
```
